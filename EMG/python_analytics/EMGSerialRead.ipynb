```python
def parse(data):
    data = data[2:-2]
    arr = data.split(" ")
    
    sync0 = arr[0]
    sync1 = arr[1]
    prot_version = arr[2]
    packet_counter = arr[3]

    ch1_high = arr[4]
    ch1_low = arr[5]
    
    ch2_high = arr[6]
    ch2_low = arr[7]

    ch3_high = arr[8]
    ch3_low = arr[9]
    
    ch4_high = arr[10]
    ch4_low = arr[11]
    
    ch5_high = arr[12]
    ch5_low = arr[13]
    
    ch6_high = arr[14]
    ch6_low = arr[15]
    
    switches_state = arr[16]
    
    print("CH1 high:", ch1_high, "low:", ch1_low)
    print("CH2 high:", ch2_high, "low:", ch2_low)
    print("CH3 high:", ch3_high, "low:", ch3_low)
    print("CH4 high:", ch4_high, "low:", ch4_low)
    print("CH5 high:", ch5_high, "low:", ch5_low)
    print("CH6 high:", ch6_high, "low:", ch6_low)
    print()

import serial
ser = serial.Serial('COM6', 57600, timeout = .1)
while True:
    data = ser.readline()[:-2] #the last bit gets rid of the new-line chars
    if data:
        parse(str(data))
```
